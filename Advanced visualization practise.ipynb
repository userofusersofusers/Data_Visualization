```python
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline 
import warnings
warnings.filterwarnings('ignore')
```

```python
# Styling dashboards
```

```python
sns.set_style('dark', {"axes.facecolor":"black"}) 
# whitegrid, darkgrid, white, dark, ticks
f, axes = plt.subplots(2, 2, figsize=(15,15))
#  Plot [0,0]
k4 = sns.kdeplot(movies.BudgetMillions, movies.AudienceRating,\
                 shade =True, shade_lowest=True, cmap='inferno',\
                 ax=axes[0,0])
k4b= sns.kdeplot(movies.BudgetMillions, movies.AudienceRating,\
                cmap='cool', ax=axes[0,0])
# Plot [0,1]
k5 = sns.kdeplot(movies.BudgetMillions, movies.CriticRating,\
                shade =True, shade_lowest=True, cmap='inferno',\
                ax=axes[0,1])
k5b= sns.kdeplot(movies.BudgetMillions, movies.CriticRating,\
                cmap='cool', ax=axes[0,1])
# plot [1,0]
z = sns.violinplot(data=movies,\
                 x='Year', y='BudgetMillions', ax=axes[1,0],\
                  palette='YlOrRd')
#zb= sns.violinplot(data=movies[movies.Genre== 'Drama'], \
 #               movies.CriticRating,\
  #              cmap='cool', ax=axes[0,0])
# Plot [1,1]
k6 = sns.kdeplot(movies.CriticRating, movies.AudienceRating, \
                shade=True, shade_lowest=False, cmap='Blues_r', \
                ax=axes[1,1])
k6b = sns.kdeplot(movies.CriticRating, movies.AudienceRating,\
                 cmap='gist_gray_r', ax=axes[1,1]) # k6b is beauty feature
k4.set(xlim=(-20,160))
k5.set(xlim=(-20,160))
```

```python
sns.set_style('whitegrid')
list1 = [] # also possible list1 = list()
mylabels=list()
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].BudgetMillions)
    mylabels.append(gen)
fig, ax = plt.subplots() # created a grid, which size is changed next line
fig.set_size_inches(11.7, 8.27) # A4paper size
h = plt.hist(list1, bins=30, stacked = True, rwidth=0.99, label=mylabels)
plt.title('Movie Budget Distribution', fontsize=35,\
         color='DarkBlue', fontname='calibri')
plt.ylabel('Number of Movies', fontsize=25, color = 'Red', \
          fontname='calibri')
plt.xlabel('Budget', fontsize=25, Color ='Red', fontname='calibri')
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.legend(frameon=True, fancybox=True, shadow=True,\
           framealpha=1, prop={'size':20})
plt.show()
```
